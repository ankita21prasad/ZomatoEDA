```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df, df_country, on = 'Country Code', how='left')
final_df.head()
```

```python
final_df.dtypes
```

```python
country_names = final_df.Country.value_counts().index
country_val = final_df.Country.value_counts()
```

# Pie Chart

```python
plt.pie(country_val[:3], labels = country_names[:3], autopct = '%.2f%%')
```

#### Zomato's maximum transactions are in India(aws it's the base country) followed by US and then UK

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
ratings
```

#### Observations -
##### Rating 4.5 to 4.9 -> Excellent
##### Rating 4.0 to 4.4 -> Very Good
##### Rating 3.5 to 3.9 -> Good
##### Rating 3.0 to 3.4 -> Average
##### Rating 2.5 to 2.9 -> Average
##### Rating 2.0 to 2.4 -> Poor

```python
ratings.head()
```

```python
import matplotlib as mlt
mlt.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings, hue='Rating color', palette = ['black', 'red', 'orange', 'yellow', 'green', 'green'])
```

### Observtion
#### 1. Not rated count very high
#### 2. Maximun ratings as 2.4 to 3.9

## Count Plot

```python
sns.countplot(x='Rating color', data=ratings, palette = ['black', 'red', 'orange', 'yellow', 'green', 'green'])
#count is frequency
```

#### Maximum zero ratings are from India
